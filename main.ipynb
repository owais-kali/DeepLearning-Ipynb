```python
#importing necessary libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0, 2, 150)
x = z * np.sin(45 * z)
y = z * np.cos(45 * z)

ax.plot3D (x, y, z, 'green')
ax.set_title('3D Line plot')
plt.show()
```
